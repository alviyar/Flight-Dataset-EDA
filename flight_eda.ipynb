```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


df=pd.read_excel('/content/flight_price.xlsx')
df.head()
```

```python
#get basic info about data
df.info()
```

```python
df.describe()
```

```python
#feature engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
#drop of date of journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
#removing the day-month from arrival time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
#arrival time
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minutes']=df['Arrival_Time'].str.split(':').str[1]
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minutes']=df['Arrival_minutes'].astype(int)
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_minutes']=df['Dep_Time'].str.split(':').str[1]
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_minutes']=df['Dep_minutes'].astype(int)
df.head()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
})
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'].fillna(df['Total_Stops'].mode()[0], inplace=True)
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_hours'] = df['Duration'].str.extract('(\d+)h').fillna(0).astype(int)
df['Duration_minutes'] = df['Duration'].str.extract('(\d+)m').fillna(0).astype(int)
df.drop('Duration',axis=1,inplace=True)
```
